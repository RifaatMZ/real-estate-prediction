Из данной таблицы нам ясно, что нужно исправить NaN в LifeSquare и Healthcare_1.
Для LifeSquare счетаем сердный процент желой площади от общей площоди и умнажаем полученную процент на общую площадь. 
Для Healthcare_1 мы меняем на медяну.

```python
living_area_avr = df1['LifeSquare'].median() / df1['Square'].median()
```

```python
living_area_avr  
```

```python
df1.loc[df1['LifeSquare'].isnull(), 'LifeSquare'] = df1['Square'] * living_area_avr
```

```python
df1.loc[df1['Healthcare_1'].isnull(), 'Healthcare_1'] = df1['Healthcare_1'].median()
```

```python
df2 = df1.drop(['Id'], axis='columns')
```

```python
df2.dtypes
```

счетаем house_age и добовляем его в наш датасет

```python
df2['house_age'] = df2['HouseYear'].apply(lambda x: 2020 - x)
```

# Visualization 

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
df2['Price'].hist()
plt.ylabel('count')
plt.xlabel('Price')

plt.subplot(122)
sns.kdeplot(df2['Price'], shade=True, legend=False)
```

```python
target_mean = round(df2['Price'].mean(), 1)
target_median = df2['Price'].median()
target_mode = df2['Price'].mode()[0]

target_mean, target_median, target_mode
```

```python
plt.figure(figsize=(12, 6))

sns.distplot(df2['Price'])

y = np.linspace(0, 0.0000075, 100)
plt.plot([target_mean] * 100, y, label='mean', linestyle=':', linewidth=4)
plt.plot([target_median] * 100, y, label='median', linestyle='--', linewidth=4)
plt.plot([target_mode] * 100, y, label='mode', linestyle='-.', linewidth=4)

plt.title('Distribution of median_house_value')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
df2['DistrictId'].hist()
plt.ylabel('count')
plt.xlabel('DistrictId')

plt.subplot(122)
sns.kdeplot(df2['DistrictId'], shade=True, legend=False)
```
